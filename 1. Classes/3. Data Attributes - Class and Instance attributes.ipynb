## Data Attributes

```python
class MyClass:
    language = 'Python'

my_obj = MyClass()
```

```python
MyClass.__dict__
```

```python
my_obj.__dict__
```

```python
# Python starts looking for language attribut in MyClass namespace

MyClass.language
```

```python
# Python starts looking in my_obj namespace
# if it finds it, returns it
# if not, it looks, in the type(class) of my_obj i.e MyClass

my_obj.language

# In this case python didn't find language attribute in my_obj namespace
# so it move to MyClass' namespace.
```

**Class Attribute ->**
'language' is a class atrribute of MyClass

**Now lets create language attribute for my_obj instance**

```python
my_obj.language = 'Ruby'
```

```python
my_obj.__dict__
```

```python
MyClass.__dict__
```

```python
my_obj.language
```

```python
# Great! We can see the value is now Ruby,
# That's because, Python looks for language attribute first in namespace
# of that object and now, it finds that attribute inside it
```

Now in this case, 

language is an **instance attribute**

___It depends upon, which namespace dict it is defined in___.

```python
# Class attribut--
print(MyClass.language)

# Instance attribute--
print(my_obj.language)
```

```python
other_obj = MyClass()
other_obj.__dict__
```

```python
other_obj.language
```

I hope now it is clear, why we got 'Python' in this case

## Example -

```python
class BankAccount:
    apr = 1.1

acc_1 = BankAccount()
acc_2 = BankAccount()
```

```python
acc_1.__dict__, acc_2.__dict__
```

```python
# They will fetch the apr from class namespace dict
acc_1.apr, acc_2.apr
```

```python
acc_1.apr = 0
acc_1.__dict__
```

```python
# We can see acc_1 fetch apr from its own namespace, whereas acc_2 still
# fetched it from Class namespace
acc_1.apr, acc_2.apr
```

**Can we create new attribute for an instance that doesn't exist in class namespace ?**

```python
acc_3 = BankAccount()

acc_3.apr = 2.4
acc_3.account_type = 'Savings'
```

```python
acc_3.__dict__, acc_2.__dict__, acc_1.__dict__
```

```python
MyClass.__dict__
```

```python
# We can see that Instance Attrb i.e account_type is only present in acc_3
```

**Yes we can create new attribute for an instance which might not be present in other instances or class**

<h3>Instances NameSpace Dict is mutable</h3>

```python
# Immutable dictionary -- we can't modify this dict directly --
type(BankAccount.__dict__)
```

```python
# we got error--
BankAccount.__dict__['debit_card'] = 'Yes'
```

```python
# mutable dictionary -- we can modify this dict directly --
type(acc_1.__dict__)
```

```python
# We did not get any error -- 
acc_1.__dict__['debit_card'] = 'Yes'
```

```python
acc_1.__dict__
```
