It gave the same thing --

But now consider following case -><br> 
    . In which programmer messed up like this (or may be intensionally)

```python
class MyClass:
    __class__ = str

m = MyClass()

type(m), m.__class__
```


we know that m is an instance of MyClass. **type()** is smarter as it returned the same which was expected, unlike --class--

```python
isinstance(m, MyClass)
```

```python
isinstance(m, str)
```

It returned true because we messed up with Dunder class method i.e --class-- in MyClass

```python
# Now it will return False

class MyClass:
    pass
m = MyClass()

isinstance(m, str)
```
