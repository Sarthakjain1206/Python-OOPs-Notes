## Data Class Attributes

```python
class Program:
    language = 'Python'
    version = '3.8'
```

```python
Program.__name__
```

```python
type(Program)
```

<h3>Get & set Class Attributes</h3>

1. By Dot Notation -

```python
print(Program.language)
print(Program.version)

# Mutate the class attributes --
Program.version = '3.6'

print("After Changing --")
print(Program.language)
print(Program.version)
```

2. By inbuilt function -- getattr() and setattr()

```python
getattr(Program, 'version')
```

```python
setattr(Program, 'version', '3.8')
```

```python
print(Program.version)
```

**Difference in Dot Notation and inbuilt function**

```python
# getattr() has default value parameter, if set to some value, then 
# it will return that value if AttributeError exception occurs, unlike
# dot-notation.

print(Program.x)
```

```python
getattr(Program, 'x')
```

```python
getattr(Program, 'x', 'Nil')
# As we can see 'x' is not an attribute of our class, still it didn't raise exception

# This functionality is not provided by Dot-Notation.
```

```python
# second parameter i.e atrr name must be a string, otherwise we will get NameError.
getattr(Program, version)
```

```python
getattr('Hello-World', 'x', 'n/a')
```

'Hello World' is an object of str class, and we tried fetching attribute 'x' from that class, which is not there, so we end up getting n/a

```python
print(type(str))
print(type(type))
print(type(int))
```

<h3> Creating/Deleting attributes at run-time </h3>

Python is a dynamic language. So, at run-time or after the class is already defined we can create or delete new attributes

```python
# All attributes present till now --
Program.__dict__

# It will return a dict(Not actuall a dictionary, it's a HashMap), which
# is only readable.
```

```python
# We can see language and version are present at this time

# Create --
Program.x = 200
setattr(Program, 'y', 20)
Program.__dict__
# New attrbs are added
```

**But keep in mind that we can do this only with our classes not with inbuilt classes**

More specifically, we can modify the states of the our own classes but not of the in-built classes.

```python
a = 'Hello'
a.x = 200
```

***Where States of these attributes are stored ?*** <br>
In general, it is stored in a **NameSpace dictionary**. And there are few exceptions to that, we will see it later on

```python
Program.__dict__

# It is of type of mappingproxy and it is read-only dict
```

```python
Program.y = 40
# we can see state has been changed
Program.__dict__
```

**Deleting of attributes**
1. By delattr() function --

```python
delattr(Program, 'y')
```

```python
Program.__dict__
```

2. By **del** keyword 

```python
del Program.x
```

```python
Program.__dict__
```

<h4> 3rd way of accessing attributes </h4>

```python
Program.__dict__['version']
```

## Callable Class Attributes

```python
class Program:
    language = 'Python'
    
    def say_hello():
        print(f"Hello From {Program.language}")
```

```python
Program.__dict__
```

```python
# Three ways to get to the function( which is a attribute)

Program.say_hello, getattr(Program, 'say_hello'), Program.__dict__['say_hello']
```

```python
# Calling this function --
Program.say_hello(), getattr(Program, 'say_hello')(), Program.__dict__['say_hello']()
```
